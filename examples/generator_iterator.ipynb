# iterator


```python
# fibonacci by iterator
def fibonacci_list(num_items):
    numbers = []
    a, b = 0, 1
    while len(numbers) < num_items:
        numbers.append(a)
        a, b = b, a+b
    return numbers
```

```python
fibonacci_list(10)
```

# generator
- next()  
  


```python
def fibonacci_gen(num_items):
    a, b = 0, 1
    while num_items:
        yield a
        a, b = b, a+b
        num_items -= 1
```

```python
gen = fibonacci_gen(10)
gen
```

```python
[next(gen) for _ in range(10)]
```

```python
# Memory Efficiency
def test_list():
    for i in fibonacci_list(100_000):
        pass

def test_gen():
    for i in fibonacci_gen(100_000):
        pass
```
